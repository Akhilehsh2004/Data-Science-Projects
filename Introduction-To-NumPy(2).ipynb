```python
# NumPy Array To Image

import numpy as np
from PIL import Image

# Create a sample NumPy array (example: 100x100 with random values)
array = np.random.randint(0, 255, (432, 575, 4), dtype=np.uint8)

# Convert the NumPy array to an image
image = Image.fromarray(array)

# Save the image to a file
image.save('dog-photo-1.jpeg')

# Show the image
image.show()
```

```python
import numpy as np
from PIL import Image

# Create a sample NumPy array (example: 432x575 with random RGBA values)
array = np.random.randint(0, 256, (432, 575, 4), dtype=np.uint8)  # Values in range [0, 255]

# Convert the NumPy array to an image
image = Image.fromarray(array, mode='RGBA')  # Specify RGBA mode explicitly

# Save the image to a file
image.save('The Legends.png')  # Use PNG for RGBA format

# Show the image
image.show()
```
