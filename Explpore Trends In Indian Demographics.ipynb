## Project 2: Explpore Trends In Indian Demographics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulated Dataset Creation
data = {
    "State": ["Uttar Pradesh", "Maharashtra", "Bihar", "West Bengal", "Madhya Pradesh"],
    "Population": [240000000, 120000000, 125000000, 100000000, 85000000],
    "Literacy Rate (%)": [67.68, 82.34, 63.82, 77.08, 70.63],
    "Population Growth Rate (%)": [20.1, 15.9, 25.4, 13.8, 18.5],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Demographic Data:")
print(df)

# Display DataFrame Information
print("\nDataFrame Information:")
print(df.info())

# Display First Few Rows
print("\nFirst Few Rows:")
print(df.head())

# Display Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# Plot 1: Population Distribution Across States
plt.figure(figsize=(10, 6))
sns.barplot(x="State", y="Population", data=df, palette="viridis")
plt.title("Population Distribution Across States", fontsize=16)
plt.xlabel("State", fontsize=12)
plt.ylabel("Population (in billions)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Literacy Rates Across States
plt.figure(figsize=(10, 6))
sns.barplot(x="State", y="Literacy Rate (%)", data=df, palette="coolwarm")
plt.title("Literacy Rates Across States", fontsize=16)
plt.xlabel("State", fontsize=12)
plt.ylabel("Literacy Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Population Growth Rates Across States
plt.figure(figsize=(10, 6))
sns.barplot(x="State", y="Population Growth Rate (%)", data=df, palette="magma")
plt.title("Population Growth Rates Across States", fontsize=16)
plt.xlabel("State", fontsize=12)
plt.ylabel("Population Growth Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Correlation Analysis
print("\nCorrelation Matrix:")
correlation = df.drop(columns=["State"]).corr()
print(correlation)

# Heatmap for Correlation
plt.figure(figsize=(8, 5))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix", fontsize=16)
plt.tight_layout()
plt.show()

# Summary Analysis
print("\nSummary Statistics:")
print(df.describe())
```
