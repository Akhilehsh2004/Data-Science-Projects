```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Sample Bollywood Movie Data (Replace with real data from IMDb or other sources)
data = {
    'movie title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E', 'Movie F', 'Movie G', 'Movie H', 'Movie I', 'Movie J'],
    'genre': ['Action', 'Romance', 'Comedy', 'Drama', 'Action', 'Comedy', 'Romance', 'Thriller', 'Drama', 'Action'],
    'release year': [2023, 2022, 2023, 2021, 2022, 2023, 2020, 2022, 2021, 2023],
    'box office crores': [150, 80, 120, 60, 200, 100, 90, 70, 50, 180],
    'lead actor': ['Actor X', 'Actor Y', 'Actor Z', 'Actor X', 'Actor Y', 'Actor Z', 'Actor X', 'Actor Y', 'Actor Z', 'Actor X']
}

# Create DataFrame
bollywood_df = pd.DataFrame(data)

# Basic Data Exploration
print("First few rows of the DataFrame:")
print(bollywood_df.head())

print("\nDataframe Information:")
print(bollywood_df.info())

# 1. Movie Count by Genre
movie_count_by_genre = bollywood_df['genre'].value_counts()
print("\nMovie Count by Genre:")
print(movie_count_by_genre)

# 2. Total Box Office Collection by Genre
total_box_office_by_genre = bollywood_df.groupby('genre')['box office crores'].sum()
print("\nTotal Box Office Collection by Genre:")
print(total_box_office_by_genre)

# 3. Total Box Office Collection by Lead Actor
total_box_office_by_actor = bollywood_df.groupby('lead actor')['box office crores'].sum()
print("\nTotal Box Office Collection by Lead Actor:")
print(total_box_office_by_actor)

# Data Visualization

# Bar Plot of Movie Count by Genre
plt.figure(figsize=(8, 6))
sns.barplot(x=movie_count_by_genre.index, y=movie_count_by_genre.values, palette='viridis')
plt.title('Movie Count by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.tight_layout()
plt.show()

# Bar Plot of Total Box Office Collection by Genre
plt.figure(figsize=(8, 6))
sns.barplot(x=total_box_office_by_genre.index, y=total_box_office_by_genre.values, palette='coolwarm')
plt.title('Total Box Office Collection by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Box Office Collection (in Crores)', fontsize=12)
plt.tight_layout()
plt.show()

# Bar Plot of Total Box Office Collection by Lead Actor
plt.figure(figsize=(8, 6))
sns.barplot(x=total_box_office_by_actor.index, y=total_box_office_by_actor.values, palette='muted')
plt.title('Total Box Office Collection by Lead Actor', fontsize=16)
plt.xlabel('Lead Actor', fontsize=12)
plt.ylabel('Box Office Collection (in Crores)', fontsize=12)
plt.tight_layout()
plt.show()

# Word Cloud for Movie Titles
plt.figure(figsize=(10, 8))
text = " ".join(bollywood_df['movie title'].values)
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Movie Titles', fontsize=16)
plt.tight_layout()
plt.show()

# Word Cloud for Movie Genres
plt.figure(figsize=(10, 8))
text = " ".join(bollywood_df['genre'].values)
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Movie Genres', fontsize=16)
plt.tight_layout()
plt.show()

# Explore trends over the years
# Group 
```
