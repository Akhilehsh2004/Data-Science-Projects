## Step-by-Step Simple Linear Regression

```python
# Step 1: Import library and function 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Create or load dataset

data = {

    'Rainfall (mm)': [500, 600, 700, 800, 900, 1000, 1100],
    'Crop Yield (Quintals/Ha)': [20, 25, 30, 35, 40, 45, 50]

}

df = pd.DataFrame(data)

```

```python
# Display dataset

df
```

```python
# Step 3: Independent variable (X) and dependent variable (y)

x = df[['Rainfall (mm)']]  # Use double brackets to maintain DataFrame structure
y = df['Crop Yield (Quintals/Ha)']

```

```python
# Step 4: Split data into training and testing sets

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2529)
```

```python
# Step 5: Intialize and fit the Linear Regression Model

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Step 6: Coefficients and Intercept

model.intercept_, model.coef_
```

```python
# Step 7: Predct values for the test set

y_preds = model.predict(x_test)
```

```python
# Display predictions

y_preds
```

```python
# Step 8: Calculate Mean Squared Error and R-squared

mean_squared_error(y_test, y_preds), r2_score(y_test, y_preds)
```

```python
# Plot actual Vs predicted values

plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_preds, color="red", linewidth=2, label='Predicted')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Crop Yield (Quintals/Ha)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
