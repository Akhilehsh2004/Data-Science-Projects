## Project 1: Analyzing IPL Data


```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data: A simplified version of IPL data
data = {
    'Season': [2019, 2019, 2020, 2020, 2021, 2021, 2022, 2022],
    'Match': [1, 2, 3, 4, 5, 6, 7, 8],
    'Team1': ['MI', 'CSK', 'RCB', 'DC', 'MI', 'RCB', 'CSK', 'KKR'],
    'Team2': ['CSK', 'RCB', 'DC', 'KKR', 'RCB', 'KKR', 'MI', 'DC'],
    'Winner': ['MI', 'CSK', 'RCB', 'DC', 'MI', 'RCB', 'CSK', 'DC'],
    'Player_of_the_Match': [
        'Hardik Pandya', 'MS Dhoni', 'Virat Kohli', 'Shikhar Dhawan',
        'Rohit Sharma', 'AB de Villiers', 'Ravindra Jadeja', 'Andre Russell'
    ],
    'Runs_Scored': [180, 175, 190, 155, 200, 185, 160, 170],
    'Home_Team': ['MI', 'CSK', 'RCB', 'DC', 'MI', 'RCB', 'CSK', 'KKR'],
    'Toss_Winner': ['MI', 'RCB', 'RCB', 'DC', 'MI', 'KKR', 'CSK', 'DC']
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print("Initial DataFrame:")
print(df.head())

# Basic info about the DataFrame
print("\nDataFrame Info:")
print(df.info())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# --------------------------------------------
# Exploratory Data Analysis (EDA)
# --------------------------------------------

# 1. Frequency of Team Wins
print("\nTeam Wins Frequency:")
print(df['Winner'].value_counts())

# 2. Player of the Match Awards
print("\nPlayer of the Match Awards Frequency:")
print(df['Player_of_the_Match'].value_counts())

# 3. Average Runs Scored
average_runs = df['Runs_Scored'].mean()
print(f"\nAverage Runs Scored: {average_runs}")

# 4. Home Advantage Analysis
df['Home_Advantage'] = df['Home_Team'] == df['Winner']
home_advantage_percentage = (df['Home_Advantage'].sum() / len(df)) * 100
print(f"\nHome Advantage Win Percentage: {home_advantage_percentage:.2f}%")

# 5. Toss Winner and Match Winner Relation
df['Toss_Winner_Match_Winner'] = df['Toss_Winner'] == df['Winner']
toss_win_match_win_percentage = (df['Toss_Winner_Match_Winner'].sum() / len(df)) * 100
print(f"\nToss Win and Match Win Percentage: {toss_win_match_win_percentage:.2f}%")

# --------------------------------------------
# Data Visualization
# --------------------------------------------

# 1. Bar Chart for Team Wins
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Winner', order=df['Winner'].value_counts().index, palette='Set2')
plt.title('Team Wins Frequency')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()

# 2. Histogram for Runs Scored
plt.figure(figsize=(8, 5))
plt.hist(df['Runs_Scored'], bins=8, color='skyblue', edgecolor='black')
plt.title('Distribution of Runs Scored')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.show()

# 3. Box Plot for Runs Scored by Season
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Season', y='Runs_Scored', palette='Set3')
plt.title('Runs Scored Across Seasons')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.show()

# 4. Scatter Plot for Runs Scored vs Match
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Match', y='Runs_Scored', hue='Season', palette='coolwarm', s=100)
plt.title('Runs Scored vs Match')
plt.xlabel('Match Number')
plt.ylabel('Runs Scored')
plt.legend(title='Season')
plt.show()

# 5. Home Advantage Visualization
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Home_Advantage', palette='Set1')
plt.title('Home Advantage Wins')
plt.xlabel('Home Advantage')
plt.ylabel('Number of Matches')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

# --------------------------------------------
# Insights
# --------------------------------------------

# Example insights that can be drawn from the data
print("\nInsights:")
print(f"- The team with the most wins: {df['Winner'].value_counts().idxmax()}")
print(f"- The player with the most 'Player of the Match' awards: {df['Player_of_the_Match'].value_counts().idxmax()}")
print(f"- Home teams won {home_advantage_percentage:.2f}% of the matches.")
print(f"- Toss winners also won {toss_win_match_win_percentage:.2f}% of the matches.")
```
