## Project 3: Examining Air Qualities In Indian Cities

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Air Quality Data for Indian Cities (Replace with real data from CPCB or OpenAQ)
data = {
    "City": ["Delhi", "Mumbai", "Kolkata", "Chennai", "Bangalore"],
    "Date": pd.date_range(start="2023-01-01", periods=5, freq="D"),
    "PM2.5": [350, 120, 200, 90, 75],
    "NO2": [50, 40, 30, 25, 20],
    "SO2": [20, 18, 15, 10, 5]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Air Quality Data:")
print(df)

# Display first few rows
print("\nFirst few rows of the DataFrame:")
print(df.head())

# Display DataFrame information (data types, non-null counts)
print("\nDataframe Information:")
print(df.info())

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Mean pollution levels across columns
print("\nMean Pollution Levels:")
print(df[["PM2.5", "NO2", "SO2"]].mean())

# City with the highest NO2 level
highest_no2_city = df.loc[df["NO2"].idxmax(), "City"]
print(f"\nCity with the highest NO2 level: {highest_no2_city}")

# Convert Date to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Plot 1: PM2.5 Trends Over Time
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["PM2.5"], marker="o", label="PM2.5", color="red")
plt.title("PM2.5 Trends Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("PM2.5 Levels", fontsize=12)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Plot 2: Compare Air Quality Across Cities
plt.figure(figsize=(10, 6))
# Group by City and calculate the mean for only numeric columns
city_data = df.groupby("City")[["PM2.5", "NO2", "SO2"]].mean()
city_data.plot(kind="bar", figsize=(10, 6), colormap="viridis")
plt.title("Average Pollution Levels Across Cities", fontsize=16)
plt.ylabel("Pollution Levels", fontsize=12)
plt.xlabel("City", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Seasonal Patterns (Simulated Monthly Data for Demonstration)
# For real seasonal data, you would want a larger dataset; this is just for demo purposes
df_monthly = df.set_index("Date").resample("M").mean()
plt.figure(figsize=(10, 6))
plt.plot(df_monthly.index, df_monthly["PM2.5"], marker="o", label="PM2.5", color="red")
plt.plot(df_monthly.index, df_monthly["NO2"], marker="s", label="NO2", color="blue")
plt.plot(df_monthly.index, df_monthly["SO2"], marker="^", label="SO2", color="green")
plt.title("Seasonal Patterns in Pollution Levels", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Levels", fontsize=12)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Heatmap for Correlations
print("\nCorrelation Matrix:")
correlation = df[["PM2.5", "NO2", "SO2"]].corr()
print(correlation)

plt.figure(figsize=(8, 5))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Air Pollutants", fontsize=16)
plt.tight_layout()
plt.show()
```
